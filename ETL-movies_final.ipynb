```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Reading IMDB_rating csv file
ratings = "Resources/IMDb_ratings.csv"
imdb_ratings_df = pd.read_csv(ratings)
imdb_ratings_df.head()
```

```python
#Reading IMDB_title csv file
imdbtitle = "Resources/IMDb_movies.csv"
imdb_titles_df = pd.read_csv(imdbtitle)
imdb_titles_df.head()
```

```python
#Reading netflix_title csv file
netflixtitles = "Resources/netflix_titles.csv"
netflix_titles_df = pd.read_csv(netflixtitles)
netflix_titles_df.head(30)
```

```python
#Reading disney_plus_show csv files
disney_plus_shows = "Resources/disney_plus_shows.csv"
disney_plus_shows_df = pd.read_csv(disney_plus_shows)
disney_plus_shows_df
```

```python
#Reducing disney_plus csv file
reduced_disney_plus_df=disney_plus_shows_df[["title","type","year","imdb_rating"]]
reduced_disney_plus_df
```

```python
#Renaming the columns of disney_plus from year to release year
renamed_disney_plus_df = reduced_disney_plus_df.rename(columns={"year":"release_year"})
renamed_disney_plus_df.head()
```

```python
#Reducing IMDB_rating csv 
reduced_imbd_rating_df=imdb_ratings_df[["imdb_title_id","weighted_average_vote"]]
reduced_imbd_rating_df
```

```python
reduced_imdb_title_df=imdb_titles_df[["imdb_title_id","title"]]
reduced_imdb_title_df
```

```python
#merging imdb_title and imdb_rating df's
merge_imdb_title_rating_df=pd.merge(reduced_imdb_title_df,reduced_imbd_rating_df, on="imdb_title_id",how="outer")
#merge_imdb_title_rating_df.lo
merge_imdb_title_rating_df=merge_imdb_title_rating_df.rename(columns={"weighted_average_vote":"imdb_ratings"})
merge_imdb_title_rating_df = merge_imdb_title_rating_df[["title","imdb_ratings"]]
print(len(merge_imdb_title_rating_df))
merge_imdb_title_rating_df = merge_imdb_title_rating_df.drop_duplicates("title")
print(len(merge_imdb_title_rating_df))
merge_imdb_title_rating_df
```

```python
#reducing netflix_title csv file
reduced_netflix_titles_df=netflix_titles_df[["title","type","release_year"]]
reduced_netflix_titles_df
```

```python
merge_netflix_ratings = pd.merge(reduced_netflix_titles_df,merge_imdb_title_rating_df, on="title",how="left")
merge_netflix_ratings
```

```python
#Merging netflix and disney df's
merge_title_df = pd.merge(merge_netflix_ratings,renamed_disney_plus_df, on="title", how="outer")
merge_title_df.loc[pd.isnull(merge_title_df["release_year_x"]),"release_year_x"] = merge_title_df["release_year_y"]
merge_title_df.loc[pd.isnull(merge_title_df["type_x"]),"type_x"] = merge_title_df["type_y"]
merge_title_df.loc[pd.isnull(merge_title_df["imdb_rating"]), "imdb_rating"] = merge_title_df["imdb_ratings"]
merge_title_df.head(8746)
```

```python
#deleting columns type_y & release_year_y from merged df
# del merge_title_df['type_y']
# merge_title_df.head()
merge_title_1_df=merge_title_df.drop(columns=['type_y', 'release_year_y',"imdb_ratings"])
merge_title_1_df
```

```python
#Renaming the column's
renamed_merge_title_1_df=merge_title_1_df.rename(columns={"type_x":"type","release_year_x":"release_year"})
merge_final_df = renamed_merge_title_1_df
```

```python
merge_final_df["netflix"] = merge_final_df["title"].isin(merge_netflix_ratings["title"])
merge_final_df["disney"] = merge_final_df["title"].isin(renamed_disney_plus_df["title"])

merge_final_df
```

```python
merge_final_df["type"].replace(to_replace = "movie", value = "Movie", inplace = True)
merge_final_df
```

```python
yr_list = []
for x in merge_final_df["release_year"]:
    year = str(x)
    year = year[:4]
    yr_list.append(year)
merge_final_df["release_year"] = yr_list
merge_final_df
```

```python
merge_final_df= merge_final_df.dropna()
merge_final_df
```

```python
imdb_rating = merge_final_df["imdb_rating"].unique()
imdb_rating = pd.DataFrame(imdb_rating)
imdb_rating = imdb_rating.rename(columns={0:"imdb_rating"})
imdb_rating = imdb_rating.dropna()
imdb_rating
```

```python
release_year = merge_final_df["release_year"].unique()
year_list = []
for x in release_year:
    year = str(x)
    year = year[:4]
    year_list.append(year)
release_year = pd.DataFrame(year_list)
release_year = release_year[0].unique()
release_year = pd.DataFrame(release_year)
release_year = release_year.rename(columns={0:"release_year"})
release_year
```

```python
# Create database connection
rds_connection_string = "postgres:password@localhost:5432/ETLproject_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
imdb_rating.to_sql(name= "imdb_rating", con = engine, if_exists = "append", index=False)
```

```python
release_year.to_sql(name= "release_year", con = engine, if_exists = "append", index=False)
```

```python
imdb_rating = pd.read_sql_query('select * from imdb_rating', con=engine)
imdb_rating
```

```python
release_year = pd.read_sql_query('select * from release_year', con=engine)
release_year
```

```python
Title = pd.merge(almost_title,imdb_rating, on="imdb_rating")

Title
```
